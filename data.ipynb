```python
import pandas as pd
import numpy as np
import random
import re
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
def unordered_text(data):
  unordered_list = []
  for i in range(len(data)):
    text = data['Ordered'][i].replace(".","")
    splited = text.split()
    random.shuffle(splited)
    un_text = ' '.join(splited)
    unordered_list.append(un_text)
  return unordered_list
```

```python
temp = unordered_text(df)
```

```python
df['Unordered'] = temp
```
